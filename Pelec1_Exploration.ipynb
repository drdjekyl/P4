### II. DataFrames harmonized

```python
import ast

# Split location column within three new Series
df_var = df2015['Location'].apply(lambda x: ast.literal_eval(x))
df2015[['Latitude', 'Longitude', 'Human_Address']] = df_var.apply(pd.Series)

# Split data in Location column and unify with df 2016
df2015.drop(labels=['Human_Address'], axis=1, inplace=True)
```

```python
# 3 columns name adjusted to fit df2015 and df2016 concatenation
df2015['TotalGHGEmissions'] = df2015['GHGEmissions(MetricTonsCO2e)']
df2015['GHGEmissionsIntensity'] = df2015['GHGEmissionsIntensity(kgCO2e/ft2)']
df2015['ZipCode'] = df2015['Zip Codes']

# Drop ancient labels
df2015.drop(labels=[
    'GHGEmissions(MetricTonsCO2e)', 'Zip Codes',
    'GHGEmissionsIntensity(kgCO2e/ft2)'
],
            axis=1,
            inplace=True)
```

```python
drop2015 = set(df2015.columns) - set(df2016.columns)
print(drop2015)
# 6 unique columns present in df2015
```

```python
drop2016 = set(df2016.columns) - set(df2015.columns)
print(drop2016)
# 1 unique columns present in df2016
```

```python
df2015.drop(labels=drop2015,
            axis=1,
            inplace=True)
# Drop meaningless columns
```

```python
df2016.drop(labels=drop2016,
            axis=1,
            inplace=True)
# Drop meaningless columns
```

```python
# Unified df with 2015 and 2016
df = pd.concat([df2015, df2016], ignore_index=True)
```

```python
# Select non residential buildings
df = df[df['BuildingType'].isin([
    'Nonresidential WA', 'Nonresidential COS', 'SPS-District K-12',
    'NonResidential'
])]

df = df.drop(columns=[
    'SecondLargestPropertyUseType', 'SecondLargestPropertyUseTypeGFA',
    'ThirdLargestPropertyUseType', 'ThirdLargestPropertyUseTypeGFA', 'ZipCode'
])
# Drop columns menaning less and with a lot of NA
```

**III.2 Feature engineering**

```python
one_hot = pd.get_dummies(df['PrimaryPropertyType'])
df.drop(columns='PrimaryPropertyType', inplace=True)
df = pd.concat([df, one_hot], axis=1)
```

```python
df['YearBuilt'] = pd.cut(df['YearBuilt'],
       11, labels=['d1', 'd2', 'd3', 'd4', 'd5', 'd6', 'd7', 'd8', 'd9', 'd10', 'd11'])

too_hot = pd.get_dummies(df['YearBuilt'])
df.drop(columns='YearBuilt', inplace=True)
df = pd.concat([df, too_hot], axis=1)
```

```python
tree_hot = pd.get_dummies(df['Neighborhood'])
df.drop(columns='Neighborhood', inplace=True)
df = pd.concat([df, tree_hot], axis=1)
```

```python
y_GHG = df['TotalGHGEmissions']
y_Cons = df['SiteEnergyUse(kBtu)']
EnergySTAR = df['ENERGYSTARScore']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Corr with building parameters
plt.figure(figsize=(9, 8))
plt.title('Location Buildings : Corr Matrix')

sns.heatmap(pd.concat([
    df[['CouncilDistrictCode', 'Latitude', 'Longitude']].astype('float64'),
    y_Cons, y_GHG
],
                      axis=1).corr(),
            annot=True,
            cbar=False,
            cmap='coolwarm');
```

```python
# Corr with GFA parameters
plt.figure(figsize=(9, 8))
plt.title('Properties: Corr Matrix')

sns.heatmap(pd.concat([
    df[[
        'LargestPropertyUseTypeGFA', 'NumberofFloors',
        'PropertyGFABuilding(s)', 'NumberofBuildings', 'PropertyGFATotal',
        'PropertyGFAParking'
    ]], y_Cons, y_GHG
],
                      axis=1).corr(),
            annot=True,
            cbar=False,
            cmap='coolwarm');
```

```python
# Corr with GFA parameters
plt.figure(figsize=(9, 8))
plt.title('Energy: Corr Matrix')

sns.heatmap(pd.concat([
    df[['SteamUse(kBtu)', 'Electricity(kBtu)', 'NaturalGas(kBtu)']],
    y_Cons, y_GHG],
    axis=1).corr(), annot=True, cbar=False, cmap='coolwarm');
```

```python
# Corr with GFA parameters
plt.figure(figsize=(9, 8))
plt.title('Primary property type: Corr Matrix')

sns.heatmap(pd.concat([df.iloc[:, 34:59], y_Cons, y_GHG], axis=1).corr(),
            cbar=True,
            cmap='coolwarm');
```

```python
# Corr with GFA parameters
plt.figure(figsize=(9, 8))
plt.title('Year Built by decades: Corr Matrix')

sns.heatmap(pd.concat([df.iloc[:, 59:70], y_Cons, y_GHG], axis=1).corr(),
            annot=True,
            cbar=False,
            cmap='coolwarm');
```

```python
# Corr with GFA parameters
plt.figure(figsize=(9, 8))
plt.title('Neighboorhood: Corr Matrix')

sns.heatmap(pd.concat([df.iloc[:, 70:], y_Cons, y_GHG], axis=1).corr(),
            annot=True,
            cbar=False,
            cmap='coolwarm');
```
